```python
import os
import shutil

def copy_first_file(source_path, destination_path):
    # Traverse through the source path
    for root, dirs, files in os.walk(source_path):
        # Check if there are no more subdirectories
        if not dirs:
            # If the current folder has files
            if files:
                # Get the name of the last subfolder
                last_subfolder_name = os.path.basename(root).replace(" ","").replace("_","").lower()
                second_last_subfolder_name = os.path.basename(os.path.dirname(root)).lower()

                # Create the new folder structure in the destination path
                new_second_last_folder_path = os.path.join(destination_path, second_last_subfolder_name)
                new_last_folder_path = os.path.join(new_second_last_folder_path, last_subfolder_name)

                os.makedirs(new_last_folder_path, exist_ok=True)

                # Copy the first file found to the new folder
                first_file_path = os.path.join(root, files[0])
                shutil.copy(first_file_path, new_last_folder_path)

                print(f"Copied {first_file_path} to {new_last_folder_path}")

# Example usage
source_path = "/4TBHD/ISL/data_preparation/seq_test_data"  # Replace with your source path
destination_path = "/4TBHD/ISL/data_preparation/check"  # Replace with your destination path

copy_first_file(source_path, destination_path)
```
